```python
#See the eldest players
eldest = df_fifa.sort_values("Age", ascending=False)[["Name", "Nationality", "Club", "Position", "Age"]].head(10)
eldest.set_index("Name", inplace=True)
print(eldest)
```

```python
#See the youngest players
youngest = df_fifa.sort_values("Age")[["Name", "Nationality", "Club", "Position", "Age"]].head(10)
youngest.set_index("Name", inplace=True)
print(youngest)
```

```python
# We want to see the players age distribution in selected clubs (Liverpool, City, Chelsea, United, Tottenham, Arsenal). Through a violin plot
# First we make a varibel string with name of the clubs
selected_clubs = ("Liverpool", "Manchester City", "Chelsea", "Manchester United", "Tottenham", "Arsenal")
# Then we assign the value from the data set
top6_clubs = df_fifa[df_fifa["Club"].isin(selected_clubs) & df_fifa["Age"]]

#Then we make a violin plot
sns.violinplot(x="Club", y="Age",data=top6_clubs)
sns.set_context("paper")
```
